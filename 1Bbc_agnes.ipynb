```python
#importing packages...
import numpy as np
from sklearn.neural_network import MLPClassifier 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import classification_report,confusion_matrix
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
#importing dataset....
data_set=pd.read_csv("bc.csv")
df=pd.DataFrame(data_set)
# print(df)
```

```python
print('A look at the data')
print(df.head(),'\n')

print ('Number of columns and rows of the data ')
print(df.shape,'\n')

print ('Description of the data')
print(df.describe(),'\n')

print ('Types of data values')
print(df.dtypes, '\n')

#identify nans
def num_missing(x):
    return sum(x.isnull())
#Applying per column:
print ("Missing values per column:")
print (df.apply(num_missing, axis=0),'\n') 
```

```python
#Converting categorical values to numerical values using label encoder
le=LabelEncoder()
# Iterating over all the common columns in train and test
for col in df.columns.values:
       # Encoding only categorical variables
    if df[col].dtypes=='object':
       # Using whole data to form an exhaustive list of levels
        data=df[col].append(df[col])
        le.fit(data.values)
        df[col]=le.transform(df[col])
# print(df)
```

```python
#Seperating the target variable
array=df.values
#the dataset without the  column: 'y'
X=array[:,2:32]
# print("X")
# print(X)
#the target column: y
Y=array[:,1]
# print("Y")
# print(Y)
```

```python
#Splitting train and test data using stratified split
X_train, X_test, Y_train, Y_test = train_test_split(X , Y, test_size=0.3, random_state = 2000,stratify=Y)
#print(Y_test.shape)
```

```python
#Applying standard scalar (Data is coverted so that mean=0 and SD=1)
scaler = StandardScaler()
scaler.fit(X_train)
scaler.fit(X_test)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Splitting data for training into training and validation sets
X_tr,X_val,Y_tr,Y_val= train_test_split(X_train,Y_train,test_size=0.2, random_state = 2000,shuffle=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier().fit(X_tr, Y_tr)
print('Accuracy of Decision Tree classifier on training set: {:.2f}'
     .format(clf.score(X_tr, Y_tr)))
print('Accuracy of Decision Tree classifier on test set: {:.2f}'
     .format(clf.score(X_val, Y_val)))
```

```python
#Different classifiers used..
classifiers = [
    KNeighborsClassifier(3),
    SVC(),
    DecisionTreeClassifier(),
    GaussianNB(),
    LinearDiscriminantAnalysis()]
```

```python
print("***After training...***")
for clf in classifiers:
    clf.fit(X_tr,Y_tr)
    
    name = clf.__class__.__name__
    print("***",name,"***")
    predictions = clf.predict(X_val)
    #Train accuracy
    ac_tr=accuracy_score(Y_tr,clf.predict(X_tr))
    print("Training accuracy:",ac_tr*100,"%")
    #Validation accuracy
   
    ac_val=accuracy_score(Y_val, predictions)
    print("Validation score:",ac_val*100,"%")
    print(confusion_matrix(Y_val,predictions))
    print(classification_report(Y_val,predictions))
```

```python
print("***After training...***")
for clf in classifiers:
    name = clf.__class__.__name__
    print("***",name,"***")
    predictions = clf.predict(X_test)

    ac_test=accuracy_score(Y_test, predictions)
    print("Testing accuracy:",ac_test*100,"%")
    print(confusion_matrix(Y_test,predictions))
    print(classification_report(Y_test,predictions))
```
